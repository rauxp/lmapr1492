```python
from pymatgen import Structure
from pymatgen import symmetry
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from pymatgen.core.operations import SymmOp
from IPython.display import Image
from IPython.display import Math

cle_api="i76sQnZ5PVPb3vayMHvN"
id_mp="mp-7988"
with MPRester(cle_api) as l:
    structure=l.get_structure_by_material_id(id_mp)
    f=CifWriter(structure)
    f.write_file('structure.cif')
    analyzer=SpacegroupAnalyzer(structure)
```

```python
print("ETUDE DE L'EFFET 3 ÉLÉMENTS DE SYMÉTRIE DIFFÉRENTS")
print(" ")
print("1ER ÉLÉMENT")
print(" ")
print("Troisième élément de symétrie (C3 axis:-y, x-y, z):")
first=SymmOp.from_xyz_string("-y, x-y, z")
print(first)
print("L'axe de rotation C3 appliqué à Cs0 #4 se trouvant en (x,y,z) = (0.67, 0.33, 0.60/1) :")
print(first.operate((0.67, 0.33, 0.60/1)))
print(" ")
Image("Cs0_C3.jpg")

```

```python
%%latex
Equation décrivant la première symétrie 

\begin{equation}

\begin{pmatrix} 0 & -1 & 0 \\1 & -1 & 0\\0 & 0 & 1\end{pmatrix}
\begin{pmatrix}x \\ y \\ z\end{pmatrix}= 
\begin{pmatrix}-y \\ x-y \\ z\end{pmatrix}

\end{equation}

\begin{equation}

\begin{pmatrix} 0 & -1 & 0 \\1 & -1 & 0\\0 & 0 & 1\end{pmatrix}
\begin{pmatrix}0.67 \\ 0.33 \\ 0.60\end{pmatrix}= 
\begin{pmatrix}-0.33 \\ 0.34 \\ 0.60\end{pmatrix}

\end{equation}
```

```python
print("/////////////////////////////////////////////")
print("2E ÉLÉMENT")
print(" ")
print("Septième élément de symétrie (C2: y, x, -z):")
second=SymmOp.from_xyz_string("y, x, -z")
print(second)
print("L'axe de rotation C2 appliqué à O1 #8 se trouvant en (x,y,z) = (0.33,0.67,0.67/1) :")
print(second.operate((0.33, 0.67, 0.67/1)))
Image("O1_C2.jpg")
```

```python
%%latex
Equation décrivant la deuxième symétrie
\begin{equation}

\begin{pmatrix} 0 & 1 & 0 \\1 & 0 & 0\\0 & 0 & -1\end{pmatrix}
\begin{pmatrix}x \\ y \\ z\end{pmatrix}= 
\begin{pmatrix}y \\ x \\ -z\end{pmatrix}

\end{equation}
\begin{equation}

\begin{pmatrix} 0 & 1 & 0 \\1 & 0 & 0\\0 & 0 & -1\end{pmatrix}
\begin{pmatrix}0.33 \\ 0.67 \\ 0.67\end{pmatrix}= 
\begin{pmatrix}0.67 \\ 0.33 \\ -0.67\end{pmatrix}

\end{equation}
```

```python
print("/////////////////////////////////////////////")
print("3E ÉLÉMENT")
print(" ")
print("Douxième élément de symétrie (plan mirroir: x, x-y, z):")
third=SymmOp.from_xyz_string("x, x-y, z")
print(third)
print("Plan miroir appliqué à CsO #19 se trouvant en (x,y,z) = (1, 1, 0.26/1) :")
print(third.operate((1, 1, 0.26/1)))
Image("CsO_mirror.jpg")
```

```python
%%latex
Equation décrivant la troisième symétrie
\begin{equation}

\begin{pmatrix} 1 & 0 & 0 \\1 & -1 & 0\\0 & 0 & 1\end{pmatrix}
\begin{pmatrix}x \\ y \\ z\end{pmatrix}= 
\begin{pmatrix}x \\ x-y \\ z\end{pmatrix}

\end{equation}
\begin{equation}

\begin{pmatrix} 1 & 0 & 0 \\1 & -1 & 0\\0 & 0 & 1\end{pmatrix}
\begin{pmatrix} 1\\ 1 \\ 0.26\end{pmatrix}= 
\begin{pmatrix}1 \\ 0 \\ 0.26\end{pmatrix}

\end{equation}
```
